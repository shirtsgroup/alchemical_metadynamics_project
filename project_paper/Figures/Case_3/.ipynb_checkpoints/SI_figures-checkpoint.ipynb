```python
#this is my code in the plumed files for defining the avaraged torsion

eta6: TORSION ATOMS=@N1-1,@C6-1,@N6-1,@C10-1
N1H61: TORSION ATOMS=@N1-1,@C6-1,@N6-1,@H61-1
N1H62: TORSION ATOMS=@N1-1,@C6-1,@N6-1,@H62-1

torsave: CUSTOM ...
  ARG=eta6,N1H62,N1H61
  VAR=a1,a2,a3
  PERIODIC=-PI,PI
  FUNC=atan2((sin(a1)+sin(a2)+sin(a3+pi))/3,(cos(a1)+cos(a2)+cos(a3+pi))/3)
...
```

```python
import math
import numpy as np

HREX=[6.07, -10.4, 1.8, 0]
HREX_err=[0.21, 0.6, 0.6, 0]


#statica bias
AM=[5.46,-9.4, 210.88-208.90,0.1]
AM_err=[0.35, 0.9, math.sqrt(0.5**2+0.19**2),0.7]

#dynamic bias, htm avg on last 100ns
AMb=[4.66,-7.9, 215.98-211.5, 1.81]
AMb_err=[0.35, 0.22, math.sqrt(0.25**2+0.15**2),0.16]
```

```python
plt.figure(figsize=(7, 6))
x=np.linspace(-11,8,50)
#plt.set_xlabel('log(Alpha)', fontsize=15)
plt.ylabel(r'$\Delta \Delta G_{AM}$ (kJ/mol)', fontsize=22)
plt.xlabel(r'$\Delta \Delta G_{HREX}$ (kJ/mol)', fontsize=22)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
names = [r'm6A_nt$_{anti/syn}$', r'm6A_dup$_{anti/syn}$', 'Duplex/nt', r'A_dup$_{anti/syn}$' ]



plt.plot(x, x, linestyle='dashed', color='k')

#plt.errorbar(HREX, HREX, HREX_err, fmt='x', elinewidth = 4, capsize=8, label='HREX')
plt.errorbar(HREX, AM, AM_err, HREX_err,  fmt='o', elinewidth = 4, capsize=8, label='Static Bias')
plt.errorbar(HREX, AMb, AMb_err, HREX_err,  fmt='o', elinewidth = 4, capsize=8, label='Dynamic Bias')
#plt.errorbar(HREX, AMc, AMc_err, HREX_err,  fmt='o', elinewidth = 4, capsize=8, label='Dynamic Bias final Bias')
plt.text(3.5, 2.6, names[0], fontsize=15)
plt.text(-11.5, -5.1, names[1], fontsize=15)
plt.text(-0.5, 6, names[2], fontsize=15)
plt.text(-1.0, -2, names[3], fontsize=15)

plt.legend(loc='lower right', fontsize=15)
```
