```python
import numpy as np
import plumed
import matplotlib.pyplot as plt
from matplotlib import rc
from mpl_toolkits import mplot3d
%matplotlib inline

rc("font", **{"family": "sans-serif", "sans-serif": ["DejaVu Sans"], "size": 10})
# Set the font used for MathJax - more on this later
rc("mathtext", **{"default": "regular"})
plt.rc("font", family="Arial")
```

```python
fes2D=plumed.read_as_pandas("duplex_fes2D.dat")
fes1D=plumed.read_as_pandas("duplex_fes1D.dat")
```

```python
#The correct 1D fes along lambda would be given by fes1D + the contribution along lambda present in fes2D. 
# The latter ca be computed as follows:
xdata=fes2D["torsave"][:100]
ydata=np.linspace(0,15,16)
zdata=np.zeros((100,16))
for i in range(16):
    for j in range(100):
        zdata[j][i]=fes2D["file.free"][i*100+j]
Glam=np.zeros(16)
for i in range(len(zdata[0])):
    # Glam[i]=2.476*np.log(np.sum(np.exp((1/2.476)*zdata[:,i])))
    Glam[i]=np.log(np.sum(np.exp((1/2.476)*zdata[:,i])))
    
plt.plot(ydata,Glam)
plt.xlabel(r'$\lambda$ state', fontsize=15)
plt.ylabel('G (kT)', fontsize=15)
plt.grid()
```

```python
#importing profile along lambda for the m6A case
m6A_pr=np.load("lambdaProfile_m6A_1D2D.npy")
```

```python
lambdas=[0.00, 0.01, 0.03, 0.05, 0.10, 0.20, 0.35, 0.45, 0.55, 0.65, 0.80, 0.90, 0.95, 0.97, 0.99, 1.00]

plt.figure(figsize=(7, 5))
plt.ylabel("G (kT)", fontsize=15)
plt.xlabel(r'$\lambda$', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.plot(lambdas,(Glam+fes1D["file.free"]-np.min(Glam+fes1D["file.free"])) / 2.494, 'o-', color='r', label="duplex")
plt.plot(lambdas,m6A_pr / 2.494, 'o-', color='b', label="m6A")
plt.legend(fontsize=13)
plt.grid()
```

```python
plt.figure(figsize=(7, 5))
plt.ylabel("G (kT)", fontsize=15)
plt.xlabel(r'$\lambda$ state', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.plot(fes1D["lambda"],(Glam+fes1D["file.free"]-np.min(Glam+fes1D["file.free"]))/2.494, 'o-', color='r', label="duplex")
plt.plot(fes1D["lambda"],m6A_pr / 2.494, 'o-', color='b', label="m6A")
plt.legend(fontsize=13)
plt.grid()
```

```python
duplex_fes_lambda_rew=np.load("lambda_profile_rew_duplex.npy")
m6A_fes_lambda_rew=np.load("lambda_profile_rew_m6A.npy")
```

```python
lambdas=[0.00, 0.01, 0.03, 0.05, 0.10, 0.20, 0.35, 0.45, 0.55, 0.65, 0.80, 0.90, 0.95, 0.97, 0.99, 1.00]

plt.figure(figsize=(7, 5))
plt.ylabel("G (kT)", fontsize=15)
plt.xlabel(r'$\lambda$', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.plot(lambdas,(duplex_fes_lambda_rew) / 2.494, 'o-', color='r', label="duplex")
plt.plot(lambdas,m6A_fes_lambda_rew / 2.494, 'o-', color='b', label="m6A")
plt.legend(fontsize=13)
plt.grid()
```

```python
plt.figure(figsize=(7, 5))
plt.ylabel("G (kT)", fontsize=15)
plt.xlabel(r'$\lambda$ state', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.plot(fes1D["lambda"],(duplex_fes_lambda_rew)/2.494, 'o-', color='r', label="duplex")
plt.plot(fes1D["lambda"],m6A_fes_lambda_rew / 2.494, 'o-', color='b', label="m6A")
plt.legend(fontsize=13)
plt.grid()
```

```python
#generate a a 2D fes subtracting for the 1D along lambda

shift_free=np.zeros(len(fes2D['lambda']))
for i in range(len(fes2D['lambda'])):
    shift_free[i]=fes2D['file.free'][i]-Glam[int(fes2D['lambda'][i])]
```

```python
plt.figure(figsize=(12,9))
ax = plt.axes()

# Data for three-dimensional scattered points
zdata =  shift_free / 2.494
xdata = fes2D["torsave"]
ydata = fes2D["lambda"]

ax.set_xlabel(r'$\eta$6-torsave (rad)', fontsize=18)
ax.set_ylabel(r'$\lambda$ state', fontsize=18)
#ax.set_xscale('log')
#ax.set_yscale('symlog', linthreshy=0.001)

ax.tick_params(axis='both', which='major', labelsize=15)

#ax.set_zscale('log')
#ax.xaxis._set_scale('log')
pnt3d=ax.scatter(xdata, ydata, c=zdata, marker='s', s=860, cmap='viridis');
#pnt3d=ax.contourf(xdata, ydata, zdata, cmap='hot')
cbar=plt.colorbar(pnt3d, aspect=20)
cbar.ax.tick_params(labelsize=15) 
cbar.set_label('G (kT)', fontsize=18)

plt.grid()
plt.show()
```

```python
#importing fes2D of the single nucleotide
fes_m6A=plumed.read_as_pandas("m6A_fes2D.dat")
```

```python
#plotting slices of the 2D profile corresponding to lambda=1
plt.figure(figsize=(8,6))

plt.xlabel(r'$\eta$6-torsave (rad)', fontsize=15)
plt.ylabel("G (kJ/mol)", fontsize=15)

plt.plot(fes_m6A["torsave"][-100:],(fes_m6A["file.free"][-100:]-np.min(fes_m6A["file.free"][-100:]))/2.494,color='b', label="m6A")
plt.plot(fes2D["torsave"][-100:],(fes2D["file.free"][-100:]-np.min(fes2D["file.free"][-100:]))/2.494,color='red', label="duplex")
plt.legend(fontsize=13)

plt.grid()
plt.show()
```

```python
import math

HREX=np.array([6.07, -10.4, 1.3])
HREX_err=np.array([0.21, 0.6, 0.6])
AM=np.array([5.46,-9.4, 210.88-208.90])
AM_err=np.array([0.35, 0.9, math.sqrt(0.5**2+0.19**2)])
```

```python
plt.figure(figsize=(7, 6))
x=np.linspace(-11,7,50)
#plt.set_xlabel('log(Alpha)', fontsize=15)
plt.ylabel(r'$\Delta \Delta G_{AM}$ (kJ/mol)', fontsize=22)
plt.xlabel(r'$\Delta \Delta G_{HREX}$ (kJ/mol)', fontsize=22)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.ylim([-8,10])
#plt.yticks([-8,-4,0,4,8])

names = [r'$m6A_{anti/syn}$', r'$Duplex_{anti/syn}$', 'Duplex/$m$6A',]



plt.plot(x, x, linestyle='dashed', color='k')

#plt.errorbar(HREX, HREX, HREX_err, fmt='x', elinewidth = 4, capsize=8, label='HREX')
plt.errorbar(HREX, AM, HREX_err, AM_err, fmt='o', elinewidth = 4, capsize=8)
plt.text(3.9, 3.1, names[0], fontsize=15)
plt.text(-12, -6.7, names[1], fontsize=15)
plt.text(1, -0.2, names[2], fontsize=15)
#handles, labels = plt.gca().get_legend_handles_labels()
#order = [0,1,2,3,4]
#plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order],loc='upper left',fontsize=15, frameon=False)
#plt.savefig('../fig4d.png', format='png', dpi=500, bbox_inches='tight')
plt.legend(loc='lower right', fontsize=15)
plt.grid()
```
